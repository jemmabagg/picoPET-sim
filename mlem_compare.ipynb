```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
its = np.array([1, 2, 3, 4])

mlem_ssim_vals = np.array([0.29433773340865216, 0.3438846836474233, 0.39145698175917454, 0.437507649422018])
mlem_cnn_ssim_vals = np.array([0.7434969641910942, 0.8268657853025443, 0.8648565377903455, 0.9196648803618467])

mlem_mse_vals = np.array([50.95611109817974,35.22302327875279, 25.987821193497496, 19.81646539451911])
mlem_cnn_mse_vals = np.array([4.316727662206719, 4.219902105228376, 2.5551487962778388, 1.6374941190307895])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # <- use fig and axes

# MSE plot
axes[0].plot(its, mlem_cnn_mse_vals, 'bo', label=f'MLEM + CNN')
axes[0].plot(its, mlem_mse_vals, 'go', label=f'MLEM')
axes[0].set_xlabel('Number of Iterations')
axes[0].set_ylabel('MSE')
axes[0].set_title('MSE vs Noise Level')

# SSIM plot
axes[1].plot(its, mlem_cnn_ssim_vals, 'bo', label=f'MLEM + CNN')
axes[1].plot(its, mlem_ssim_vals, 'go', label=f'MLEM')
axes[1].set_xlabel('Number of Iterations')
axes[1].set_ylabel('SSIM')
axes[1].set_title('SSIM vs Noise Level')

# Unified legend above both plots
handles, labels = axes[1].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=2, fontsize=10)

plt.tight_layout(rect=[0, 0, 1, 0.93])  # Leave room for the legend
plt.show()
```
